```python
import numpy as np
import pandas as pd
import random
import time
```

```python
def sqdist(a, b):
    return np.sum((a - b) ** 2)
```

```python
def readVectorsSeq(filename):
    """
    Reads comma-separated vectors from file
    Returns list of numpy arrays
    """
    data = []
    
    with open(filename, 'r') as f:
        for line in f:
            values = list(map(float, line.strip().split(',')))
            data.append(np.array(values))
    
    return data
```

```python
def kcenter(P, k):
    """
    Selects k centers using Farthest First Traversal
    P = list of points
    k = number of centers
    """
    centers = []

    centers.append(random.choice(P))

    while len(centers) < k:
        farthest_point = None
        max_dist = -1

        for p in P:
            # distance to nearest current center
            min_dist = min(sqdist(p, c) for c in centers)

            if min_dist > max_dist:
                max_dist = min_dist
                farthest_point = p

        centers.append(farthest_point)

    return centers
```

```python
def kmeansPP(P, k):
    """
    Selects k centers using k-means++ initialization
    """
    centers = []

    # First center randomly
    centers.append(random.choice(P))

    while len(centers) < k:
        # Compute squared distance to nearest center
        distances = []
        for p in P:
            d = min(sqdist(p, c) for c in centers)
            distances.append(d)

        total = sum(distances)

        # Probability distribution
        probs = [d / total for d in distances]

        # Select next center based on weighted probability
        idx = np.random.choice(len(P), p=probs)
        centers.append(P[idx])

    return centers
```

```python
print("kmeansPP function loaded")
```

```python
def kmeansObj(P, C):
    """Computeing average squared distance of each point to its nearest center"""
    total = 0

    for p in P:
        min_dist = min(sqdist(p, c) for c in C)
        total += min_dist

    return total / len(P)
```

```python
filename = r".\Assignment 4- datasets\Q1- UCI Spam clustering\spambase.data"
P = readVectorsSeq(filename)

print("Total points:", len(P))
print("Dimensions:", len(P[0]))
```

```python
"""Setting parameters"""
k = 10
k1 = 30

"""kcenter Runtime"""
start = time.time()
C1 = kcenter(P, k)
end = time.time()

print("kcenter completed")
print("Time taken:", round(end - start, 4), "seconds")
```

```python
"""Kmeans++ objective """
C2 = kmeansPP(P, k)
obj1 = kmeansObj(P, C2)

print("kmeans++ objective value:", round(obj1, 4))
```

```python
"""Corset Method"""

X = kcenter(P, k1)
C3 = kmeansPP(X, k)

obj2 = kmeansObj(P, C3)
print("Coreset method objective value:", round(obj2, 4))
```
